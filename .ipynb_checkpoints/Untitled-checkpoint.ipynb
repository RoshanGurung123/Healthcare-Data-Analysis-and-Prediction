```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('diabetic_data.csv')
df.head()
```

```python
# Since the empty data in the data set are '?' value replacing the ? value with NAn 
df.replace('?', np.nan, inplace=True)
```

```python
# Check for null values in the entire DataFrame
null_values = df.isnull()

# Count the number of null values in each column
null_count = df.isnull().sum()

# Display the null count for each column
# print(null_count)

# Calculate the percentage of null values in each column
null_percentage = (df.isnull().sum() / len(df)) * 100

# Display the null percentage for each column
print(null_percentage)
```

```python
# dropping the columns with mor enumber of empty data
df=df.drop(['weight','payer_code','medical_specialty'], axis=1)
```

```python
df.columns
```

```python
# Iterate through each column in the DataFrame
for column in df.columns:
    # Check if the number of unique values in the column is 1
    if df[column].nunique() == 1:
        # Print the column name and value for reference
        print(f"Column '{column}' has only one distinct value: {df[column].unique()[0]}")
        # Drop the column from the DataFrame
        df.drop(column, axis=1, inplace=True)
```

```python
# seperating the categorical and numerical features
# Seperating the numeric and categorical dataset
numeric_columns = final_df.select_dtypes(include=['int64']).columns.tolist()
categorical_columns = final_df.select_dtypes(include=['object']).columns.tolist()
```

```python
# data encoding
# encoding for readmitted feature
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode the 'readmitted' variable
final_df['readmitted'] = label_encoder.fit_transform(final_df['readmitted'])

# Check the mapping of encoded values to original values
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Mapping of encoded values:")
print(mapping)
```

```python
# Encoding for meformin binary encoding
# Define the mapping for binary encoding
mapping = {'No': 0, 'Steady': 1, 'Up': 1, 'Down': 1, '>7':1,'>8':1,'Norm':0,'None':-99,'>200':1,'>300':1}

# Apply the mapping to the "metformin" column
final_df['metformin'] = final_df['metformin'].map(mapping)
final_df['insulin']=final_df['insulin'].map(mapping)
final_df['A1Cresult'] = final_df['A1Cresult'].map(mapping)
final_df['max_glu_serum'] = final_df['max_glu_serum'].map(mapping)
```

```python
# Binary encoding for diabetesMed
# Define the mapping for binary encoding for diabetesMed, change, gender
mapping = {'No': 0, 'Yes': 1,'Male':1,'Female':0,'Unknown/Invalid':-99,'Ch':1,'No':0}

# Apply the mapping to the "metformin" column
final_df['diabetesMed'] = final_df['diabetesMed'].map(mapping)
final_df['gender'] = final_df['gender'].map(mapping)
final_df['change'] = final_df['change'].map(mapping)
```

```python
# Binary encoding for diabetesMed
# Define the mapping for binary encoding for race
mapping = {'Caucasian': 0, 'AfricanAmerican': 1,'Hispanic':2,'Other':3,'Asian':4}

# Apply the mapping to the "metformin" column
final_df['race'] = final_df['race'].map(mapping)
```

```python
final_df['race'].value_counts()
```

```python
# Since the age value is in midpoint so in order to train the data in the model converting the age to numeric value by taking the modpoint of the data
# Define a function to extract numeric values from the age ranges

def extract_midpoint(age_range):
    match = re.search(r'(\d+)-(\d+)', age_range)
    if match:
        start = int(match.group(1))
        end = int(match.group(2))
        midpoint = (start + end) / 2
        return midpoint
    return None

# Apply the function to extract and calculate the midpoint
final_df['age'] = final_df['age'].apply(extract_midpoint)
```

```python
# Calculate and display skewness for each numeric feature
for feature in numeric_columns:
    skewness = final_df[feature].skew()
    print(f"Skewness for '{feature}': {skewness:.2f}")
```

```python
# Calculate skewness for each numeric feature
skewness_values = [final_df[feature].skew() for feature in numeric_columns]

# Create a bar chart to visualize skewness values
plt.figure(figsize=(10, 6))
plt.barh(numeric_columns, skewness_values, color='skyblue')
plt.xlabel('Skewness')
plt.title('Skewness of Numeric Features')
plt.gca().invert_yaxis()  # Invert the y-axis to display features from top to bottom
plt.show()

# Create histograms to visualize the distribution of each numeric feature
for feature in numeric_columns:
    plt.figure(figsize=(6, 4))
    plt.hist(final_df[feature], bins=20, color='skyblue')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Since all of the features have skewed value so to reduce the skewness i am using squared root
# Apply the square root transformation to all numeric columns
for column in numeric_columns:
    final_df[column] = np.sqrt(final_df[column])
```
